## Data Analysis From Yahoo Finance

Import necessary dependencies and load the data. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks = pd.read_csv('results.csv')
```

```python
stocks.head(5)
```

stocks.select('company').show(5)

### Visualization 1: Change in high price per hour for each company

```python
sns.catplot(data=stocks, kind="bar", x="hour", y="hourly_high", hue="company", height = 7)
```

### Visualization 2: Show stocks highest price of the day.

```python
highest_price = stocks.groupby('company').max()['hourly_high'].to_frame()
sns.relplot(data=highest_price, kind="line", height = 7)
```
